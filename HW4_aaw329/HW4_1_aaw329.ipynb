Objective: create plots relative sample mean and distribution/population size for various types of distributions



```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Use this mean for all distributions
mean = 100

# 100 evenly-spaced population sizes (as integers) from 10 to 2000
num_pops = 100
pop_sizes = N_array = np.linspace(10, 2000, num_pops, dtype=int)  
#pop_sizes = (2000 / (np.array(range(1, 100)))).astype(int)

# Create a dictionary that will hold each distribution type we're going to work with
dist_dict = {}

# Create a dictionary that will hold the means of each distribution
mean_dict = {}

# Use fixed seed for np.random
np.random.seed(123)
```

## Normal distribution

```python
dist_name = "normal"
dist_dict[dist_name] = []
mean_dict[dist_name] = []

# Create a distribution for every population size in pop_sizes
for i in range(num_pops):
    N = pop_sizes[i]
    dist_dict[dist_name].append(np.random.normal(mean, size=N))
    
# Record the mean for each distribution 
for i in range(num_pops):
    mean_dict[dist_name].append(dist_dict[dist_name][i].mean())

# Plot the means!
normal_plot = plt.figure(figsize=(10,6)).add_subplot(111)
for i in range(num_pops):
    normal_plot.plot(pop_sizes[i], mean_dict[dist_name][i], "o")

# Add labels to the plot
normal_plot.set_xlabel("Population size", fontsize=14)
normal_plot.set_ylabel("Population mean", fontsize=14)
normal_plot.set_title("%s distribution means" % dist_name.title(), fontsize=18)

# Draw a dotted line to show the chosen mean
normal_plot.plot([pop_sizes[0], pop_sizes[num_pops - 1]], [mean, mean], "k--")
```

#### Figure 1: normal distributions of varying sample size. Adhering to the central limit theorem, as sample/population size increases, so the means get closer to the expected central value

## Poisson distribution

```python
dist_name = "poisson"
dist_dict[dist_name] = []
mean_dict[dist_name] = []

# Create a distribution for every population size in pop_sizes
for i in range(num_pops):
    N = pop_sizes[i]
    dist_dict[dist_name].append(np.random.poisson(mean, size=N))
    
# Record the mean for each distribution 
for i in range(num_pops):
    mean_dict[dist_name].append(dist_dict[dist_name][i].mean())

# Plot the means!
poisson_plot = plt.figure(figsize=(10,6)).add_subplot(111)
for i in range(num_pops):
    poisson_plot.plot(pop_sizes[i], mean_dict[dist_name][i], "o")

# Add labels to the plot
poisson_plot.set_xlabel("Population size", fontsize=14)
poisson_plot.set_ylabel("Population mean", fontsize=14)
poisson_plot.set_title("%s distribution means" % dist_name.title(), fontsize=18)

# Draw a dotted line to show the chosen mean
poisson_plot.plot([pop_sizes[0], pop_sizes[num_pops - 1]], [mean, mean], "k--")
```

#### Figure 2: Poisson distributions of varying sample size. Adhering to the central limit theorem, as sample/population size increases, so the means get closer to the expected central value

## Binomial distribution

```python
dist_name = "binomial"
dist_dict[dist_name] = []
mean_dict[dist_name] = []

# Create a distribution for every population size in pop_sizes
for i in range(num_pops):
    N = pop_sizes[i]
    dist_dict[dist_name].append(np.random.binomial(200, 0.5, size=N))
    
# Record the mean for each distribution 
for i in range(num_pops):
    mean_dict[dist_name].append(dist_dict[dist_name][i].mean())

# Plot the means!
binomial_plot = plt.figure(figsize=(10,6)).add_subplot(111)
for i in range(num_pops):
    binomial_plot.plot(pop_sizes[i], mean_dict[dist_name][i], "o")

# Add labels to the plot
binomial_plot.set_xlabel("Population size", fontsize=14)
binomial_plot.set_ylabel("Population mean", fontsize=14)
binomial_plot.set_title("%s distribution means" % dist_name.title(), fontsize=18)

# Draw a dotted line to show the chosen mean
binomial_plot.plot([pop_sizes[0], pop_sizes[num_pops - 1]], [mean, mean], "k--")
```

#### Figure 3: binomial distributions of varying sample size. Adhering to the central limit theorem, as sample/population size increases, so the means get closer to the expected central value

## Chi-squared distribution

```python
dist_name = "chi-squared"
dist_dict[dist_name] = []
mean_dict[dist_name] = []

# Create a distribution for every population size in pop_sizes
for i in range(num_pops):
    N = pop_sizes[i]
    dist_dict[dist_name].append(np.random.chisquare(mean, size=N))
    
# Record the mean for each distribution 
for i in range(num_pops):
    mean_dict[dist_name].append(dist_dict[dist_name][i].mean())

# Plot the means!
chi_squared_plot = plt.figure(figsize=(10,6)).add_subplot(111)
for i in range(num_pops):
    chi_squared_plot.plot(pop_sizes[i], mean_dict[dist_name][i], "o")

# Add labels to the plot
chi_squared_plot.set_xlabel("Population size", fontsize=14)
chi_squared_plot.set_ylabel("Population mean", fontsize=14)
chi_squared_plot.set_title("%s distribution means" % dist_name.title(), fontsize=18)

# Draw a dotted line to show the chosen mean
chi_squared_plot.plot([pop_sizes[0], pop_sizes[num_pops - 1]], [mean, mean], "k--")
```

#### Figure 4: chi-squared distributions of varying sample size. Adhering to the central limit theorem, as sample/population size increases, so the means get closer to the expected central value

## Geometric distribution

```python
dist_name = "geometric"
dist_dict[dist_name] = []
mean_dict[dist_name] = []

# Create a distribution for every population size in pop_sizes
for i in range(num_pops):
    N = pop_sizes[i]
    dist_dict[dist_name].append(np.random.geometric((1.0/100.0), size=N))  # The mean is 1/p
    
# Record the mean for each distribution 
for i in range(num_pops):
    mean_dict[dist_name].append(dist_dict[dist_name][i].mean())

# Plot the means!
geometric_plot = plt.figure(figsize=(10,6)).add_subplot(111)
for i in range(num_pops):
    geometric_plot.plot(pop_sizes[i], mean_dict[dist_name][i], "o")

# Add labels to the plot
geometric_plot.set_xlabel("Population size", fontsize=14)
geometric_plot.set_ylabel("Population mean", fontsize=14)
geometric_plot.set_title("%s distribution means" % dist_name.title(), fontsize=18)

# Draw a dotted line to show the chosen mean
geometric_plot.plot([pop_sizes[0], pop_sizes[num_pops - 1]], [mean, mean], "k--")
```

#### Figure 5: geometric distributions of varying sample size. Adhering to the central limit theorem, as sample/population size increases, so the means get closer to the expected central value

```python
import pandas as pd
%matplotlib inline
```

```python
# Get yearly fatalities from Vision Zero
# http://www.nyc.gov/html/dot/html/about/vz_datafeeds.shtml
url = "http://www.nyc.gov/html/dot/downloads/misc/fatality_yearly.csv"
fatalities = pd.read_csv("http://www.nyc.gov/html/dot/downloads/misc/fatality_yearly.csv")
```

```python
# Count number of fatalities for every year
yearly_fat = fatalities.groupby(["YR"])["Fatalities"].count()
```

### Make a plot, and try to make it good

```python
ax = yearly_fat.plot.bar(figsize=(12,8), fontsize=14, zorder=2)

# Set title and labels for axes 
ax.set_ylabel("Fatalities", fontsize=18)
ax.set_xlabel("Year", fontsize=18)
ax.set_title("Vision Zero: NYC traffic crash fatalities per year 2009-2018", fontsize=20)

# Remove borders around the bars since they're unnecessary
ax.spines["top"].set_visible(False)
ax.spines["left"].set_visible(False)
ax.spines["right"].set_visible(False)
ax.spines["bottom"].set_visible(False)

# Add horizontal lines to increase ease of comparison between years
ax.grid(True, axis="y", zorder=0)
```

#### Figure 1: number of fatalities reported by Vision Zero per year for the past 9 years. There is an overall downward trend, but there are still many fatalities occurring every year. Note that data for 2018 is, of course, incomplete.
